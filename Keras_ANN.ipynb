```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Downloads/loan.csv')
ds = df.copy()
ds.head()
```

```python
ds.isna().sum()
```

```python
ds.drop('Loan_ID', axis=1, inplace=True)
```

```python
ds.dropna(subset=['Gender', 'Dependents','Self_Employed', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History'], axis = 0, inplace=True)
```

```python
ds.isna().sum()
```

```python
ds.describe()
```

```python
ds.sample(10)
```

```python
d = []
for i in ds.columns:
    if i not in ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']:
        print(i,": ",ds[i].unique())
        d.append(i)
```

```python
print(d)
```

Continous variable: ApplicantIncome, CoapplicantIncome, LoanAmount

Categorical Variable: Gender, Married, Dependents, Education, Self_Employed, Loan_Amount_Term, Credit_History, Property_Area, Loan_status

```python
cn_col = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']
ct_col = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed',
                      'Loan_Amount_Term', 'Credit_History', 'Property_Area']
```

```python
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, MinMaxScaler
from itertools import chain
# Passing continous variables through scaling process
scaler = MinMaxScaler(feature_range=(0,1))
cn = scaler.fit_transform(ds[cn_col])

# Passing Cotegorical variables through Label encoder
enc = OrdinalEncoder()
ct = enc.fit_transform(ds[ct_col])

# Passing y variable to label encoder
lb_enc = LabelEncoder()
ds['Loan_Status'] = lb_enc.fit_transform(ds['Loan_Status'])
```

```python
ds[cn_col] = cn
ds[ct_col] = ct
ds.sample(10)
```

```python
ds.head(10)
```

```python
x = ds.drop('Loan_Status', axis=1).values
y = ds['Loan_Status'].values
print(type(x),'--------', type(y))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.5)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Model architecture
model = keras.Sequential([
    layers.Dense(units = 11,input_shape=(11,), activation = 'relu'),
    layers.Dense(units = 22, activation = 'relu'),
    layers.Dense(units = 1, activation = 'sigmoid') 
])
```

```python
#model.summary()
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics='accuracy')
```

```python
model.fit(x = x_train,
          y = y_train, 
          batch_size= 10, 
          epochs = 50, 
          validation_split= 0.1, 
          shuffle = True,
          verbose = 2
          
         )
```

```python
pred_y = model.predict(x_test)
```

```python
y_pred = (pred_y>0.5)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
